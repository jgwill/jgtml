

### Simplifying the Target Variable

To make the prediction task binary (profitable or not), we can transform the fdb_target into a binary variable:

1 if fdb_target > 0 (Profitable)  
0 if fdb_target <= 0 (Not Profitable)

#### Next Steps:

Transform the fdb_target into a binary variable.

Split the data into training and testing sets.

Train a classification model.

Evaluate the model.




```python
data
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Transform the target variable into a binary variable
data['fdb_target_binary'] = (data['fdb_target'] > 0).astype(int)

# Select features and target
features = data.drop(columns=[ 'fdb_target', 'fdb_target_binary', 'fdb'])
target = data['fdb_target_binary']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

```

Alternative Approach:
We can manually oversample the minority class by duplicating its instances in the training set. Let's proceed with this approach.




```python

# Manually oversample the minority class in the training set
minority_class = X_train[y_train == 1]
minority_target = y_train[y_train == 1]

# Duplicate the minority class instances
X_train_oversampled = pd.concat([X_train, minority_class] * 10, axis=0)
y_train_oversampled = pd.concat([y_train, minority_target] * 10, axis=0)

# Train the RandomForestClassifier on the oversampled data
model_oversampled = RandomForestClassifier(random_state=42)
model_oversampled.fit(X_train_oversampled, y_train_oversampled)

# Predict on the test set
y_pred_oversampled = model_oversampled.predict(X_test)

# Evaluate the model
accuracy_oversampled = accuracy_score(y_test, y_pred_oversampled)
report_oversampled = classification_report(y_test, y_pred_oversampled)

accuracy_oversampled

```
