```python
figs={}
#foreach object in pc
for k,v in pc.items():
    print(k)
    for tf in rq.timeframes:
        print(tf)
        _p=v[tf]
        tst=_p
        _idx = k + "_" + tf
        figs[_idx] = tst["fig"]
```

```python

import matplotlib.pyplot as plt
```

```python
figures= figs
```

```python

num_figures = len(figs)
num_cols = num_figures # Set the number of columns equal to the number of figures
fig, axes = plt.subplots(1, num_cols, figsize=(num_cols * 6, 6))


# Iterate through the figures and plot them
for i, (title, figure) in enumerate(figures.items()):
    ax = axes[i]
    ax.set_title(title)
    ax.axis('off')
    ax.imshow(figure.figure)  # Replace this with the appropriate method to display the figure

plt.show()
```
